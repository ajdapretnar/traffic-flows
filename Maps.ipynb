```python
import folium
import pandas as pd
import branca.colormap as cm
```

```python
df = pd.read_csv("counters-motorbikes-averages.csv")
```

```python
sep = df[df['Month'] == 9]
sep = sep.dropna()
sep.head()
```

```python
m = folium.Map(location=[46.119944, 14.815333], zoom_start=9)  # GEO center
```

```python
#create a color map
color_var = 'S1: Motor-MO' #what variable will determine the color
linear = cm.LinearColormap(
    ['green', 'yellow', 'red'],
    vmin=sep[color_var].min(), 
    vmax=sep[color_var].max(),
)
```

```python
# scale the data for size
radius = (sep[color_var]-sep[color_var].min())/(sep[color_var].max()-sep[color_var].min())
```

```python
for row in sep.iterrows():
    stm = row[1]['STM']
    traffic = row[1]['S1: Motor-MO']
    lat = row[1]['Lat']
    lon = row[1]['Lon']
    folium.CircleMarker(location=[lat, lon],
                        # we need to enhance the differences in size
                        radius=(1+radius.loc[row[0]])**3, 
                        color=linear(row[1][color_var]),
                        fill=True, 
                        fill_color=linear(row[1][color_var]),
                        popup=stm).add_to(m)
```
